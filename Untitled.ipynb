The purpose of this project is to analyze player data to identify characteristics about players and to be able to market to them better. By examining player demographics, gameplay activity, and time of play, we can identify patterns within the data that will allow us to determine things about the players. The analysis is based upon two datasets: one containing information about players and another recording individual sessions logged on the Minecraft server. Together, these provide both demographic data and behavioral information about player activity.
The players dataset includes 196 entries with variables of experience level, whether or not a player is subscribed to a newsletter about the game, their email, the amount of hours they have logged on the server, their names, genders, and finally ages. On average, players have logged 5.85 hours of playtime and have a mean age of 21.14 years. Of the 196 players, 144 are subscribed to the newsletter. The dataset contains 127 male players, 37 female players, 15 non-binary players, 11 preferring not to say, six Two-Spirit players, two Agender players, and one who selected "Other." Age and hours played are the only quantitative variables, newsletter is a binary variable, and the remaining fields are all categorical. 
The sessions dataset contains 1,535 entries. It has five variables; quantitative start and end times for each session and categorical dates of these and emails of the players. 
While the available data is sufficient for some predictive modeling, many potentially useful features are missing. Information such as personality traits, how often players actually open the newsletter, and geographic location of players could provide additional context for understanding players interests within the server. 
The question that I am attempting to answer generally is the first given question: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types? More specifically, my project aims to determine the following: Which combination of gender, age, and experience level best predicts whether or not a player subscribes to the newsletter?
To address this question, a k-nearest neighbors (KNN) classification model will be implemented. KNN is appropriate because the target variable (newsletter subscription) is categorical, and the model has multiple predictors, making a KNN model applicable. The quantitative predictors will be scaled and centered, and the data will be split into 70% for the training set and 30% for the testing set. Cross-validation will then be used to identify the optimal value of k. Once the model is trained, we will use the testing data to determine if our model is good.
Although this model may not capture every factor influencing subscription behavior, it will allow us to determine if these variables have any relationship in predicting subscription. Future analyses could incorporate more differentiated data to improve predictions.
